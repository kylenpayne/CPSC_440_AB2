```r
y <- rnorm(100, 0, sd=1)
B <- rep(1:2, each=50)
A <- rep(1:2, rep=50)
```

```r
data <- data.frame(y=y, B=B, A=A)
```

```r
data[which(data$B==1), 'y'] <- data[which(data$B==1), 'y'] + rnorm(50, 7, sd=0.1)
data[which(data$B==2), 'y'] <- data[which(data$B==2), 'y'] + rnorm(50, 14, sd=0.1)
data[which(data$A==1), 'y'] <- data[which(data$A==1), 'y'] + rnorm(50, 0, sd=1)
data[which(data$A==2), 'y'] <- data[which(data$A==2), 'y'] + rnorm(50, 0, sd=1)
```

```r
data$A<-factor(data$A)
data$B<-factor(data$B)
```

```r
interaction.plot(x.factor = data$B,trace.factor = data$A,response = data$y)
```

```r
y <- rnorm(150, 0, sd=1)
B <- rep(1:2, each=75)
A <- rep(1:3, rep=50)
data <- data.frame(y=y, B=B, A=A)
data$A<-factor(data$A)
data$B<-factor(data$B)

data[which(data$B==1), 'y'] <- data[which(data$B==1), 'y'] + rnorm(75, 7, sd=0.1)
data[which(data$B==2), 'y'] <- data[which(data$B==2), 'y'] + rnorm(75, 14, sd=0.1)
data[which(data$A==1), 'y'] <- data[which(data$A==1), 'y'] + rnorm(50, 1, sd=0.1)
data[which(data$A==2), 'y'] <- data[which(data$A==2), 'y'] + rnorm(50, -1, sd=0.1)
data[which(data$A==3), 'y'] <- data[which(data$A==3), 'y'] + rnorm(50, 5, sd=0.1)
data[which(data$A==3 & data$B==2), 'y'] <- data[which(data$A==3 & data$B==2), 'y'] - 6

interaction.plot(x.factor = data$B,trace.factor = data$A,response = data$y)
```
